```
class Node:
  def __init__(self, item = None):
    self.item = item
    self.link = None
    
class Linked_List:
  def __init__(self):
    self.root = Node()
    
  def append(self,item):
    curNode = self.root
    NewNode = Node(item)  
    if curNode.item == None:
      self.root = Node(item)
    else:   
      while curNode.link != None:
        curNode = curNode.link
      curNode.link = NewNode
      
  def print(self):
    curNode = self.root
    while curNode != None:
      print(curNode.item)
      curNode = curNode.link
  
  def delete(self, item):
    delNode = Node(item) 
    curNode = self.root
    if delNode.item == curNode.item:
      self.root = self.root.link          
    else:
      while delNode.item != curNode.item:
        preNode = curNode
        curNode = curNode.link
      preNode.link = curNode.link
  
  def size(self):
    curNode = self.root
    if curNode.item == None:
      print('0')
    else:
      size = 1
      while curNode.link != None:
        size += 1
        curNode = curNode.link
      print(size, "개의 element가 있습니다")
        
  def insert(self,item, index):
    curNode = self.root
    newNode = Node(item)
    count = 0
    if index == count:
      self.root = newNode
      newNode.link = curNode
    else:
      while count != index:
        preNode = curNode
        curNode = curNode.link
        count += 1
      preNode.link = newNode
      newNode.link = curNode
  
  def find(self, item):
    curNode = self.root    
    if curNode.item == None:
      print('-1')
    else:
      index = -1
      while curNode.link != None:
        index += 1
        if curNode.item == item:
          print(index, "번째 위치해있습니다")
          break
        curNode = curNode.link
      return ("마지막에 위치해있습니다")

     
  
  
  
```
